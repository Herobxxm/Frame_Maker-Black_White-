2. Import ส่วนต่างๆที่จะใช้

```python
import cv2
import numpy as np
import PIL
```

3. Save วิดีโอเป็น รูปแบบ frame by frame

```python
vidcap = cv2.VideoCapture('bad_apple.mp4')
success,image = vidcap.read()
count = 0
while success:
  cv2.imwrite("frame%d.jpg" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success) # Status
  count += 1
print(count) #check quantity of frame
```



4. นำแต่ละ frame มาแปลงเป็นภาพขาวดำ แล้ว ยัดใส่ไฟล์ bad_a.txt(ช้านิดนึง)
> ภาพความละเอียด 360 x 480


```python
Bad_A = open("bad_a.txt","w") # file ที่เราจะบันทึก
for i in range(0,6570):  # 6570 เป็น จำนวน Frame ทั้งหมด โดยvideoนี้ใช้เวลา 239 Sec ใช้ Frame ทั้งหมด 6570 โดยประมาณ
  File = "frame%d.jpg" %i
  Pic = PIL.Image.open(File)
  
  ## ส่วนนี้ทำหน้าที่ เปลี่ยน .jpg เป็น array มีมิติเป็น <360,480>
  
  Data_BW = Pic.convert('L')
  bw = np.asarray(Data_BW).copy()
  bw[bw < 128] = 0    
  bw[bw >= 128] = 255

########################################################

  for j in range(0,40): # เนื่องจาก Display ของเรา มีแค่ 40 x 160 เท่านั้น เราจึงจำเป็นต้องย่อ ขนาดภาพลง
    line = ""
    x = j*9
    for k in range(0,160): #3
      y = k*3
      pixel = 0
      for a in range(0,9):
        for b in range(0,3):
          pixel += bw[x+a][y+b]
      pixel /= 27
      if(pixel < 128):
        line += " 0"
      else:
        line += " 1"
######################################################3
    Bad_A.write(line) # บันทึกลงใน file .txt ที่เตรียมไว้
    Bad_A.write("\n") 
  Bad_A.write("nextFrame")
  Bad_A.write("\n")
  print("Frame %d save successfully" %i)
Bad_A.close()
```
